```python
print ('Using Pandas to read the file.\n')
print ('Displaying the first three records along with the column header')
readcsv = pd.read_csv(fname) # reading the casv file using pandas
readcsv.head (n = 10) # Displaying the first three records along with the column header
```

```python
print ('Problem 2 Section 5\n')
print ('The size of the housing dataset. \n')
size_housing = readcsv.size # Size of a dataframe is the number of fields in the dataframe i.e number of rows * number of columns
print (size_housing)
```

```python
print ('Problem 2 Section 6\n')
print ('Using Numpy to read the file.\n')
csv = np.genfromtxt(fname,  dtype=None,  names=True, delimiter=",", encoding=None) # reading the casv file using NumPy
print(csv.dtype) # displaying the data types for the columns
```

```python
print ('Problem 2 Section 6\n')
print ('Following are the data typs for the columns. \n')
print (readcsv.dtypes)# displaying the data types for the columns
```

```python
print ('Problem 2 Section 7\n')
print ('The dimensions of the housing dataset. \n')
dim_housing = readcsv.shape # Returns number of rows and the number of columns in the DataFrame
print ('Dataframe shape :', dim_housing)
print ('\nNumber of rows :', dim_housing[0]) # Returns number of rows in the DataFrame
print ('\nNumber of columns :', dim_housing[1]) # Returns number of columns in the DataFrame
```

```python
print ('Problem 2 Section 8\n')
print ('Histogram of population \n')
plt.hist(readcsv['population'], bins='auto', Color = 'blue') # Reading the population column from csv file. Setting the number of bins to Auto
plt.ylabel('Frequency') # Naming Y axis
plt.xlabel('Bins') # Naming X axis
plt.title('Population') # Title
plt.show() # Displaying plot
```

```python
print ('Problem 2 Section 9\n')
print ('Bar Chart for Ocean Proximity \n')
ax = readcsv.groupby('ocean_proximity')['ocean_proximity'].count().plot(kind = 'bar', legend=True, Color = 'red') # Grouping by based on ocean_proximity and taking count of records
plt.xlabel("Different Ocean Proximity") # Naming X axis
plt.ylabel("Frequency") # Naming Y axis
plt.title('Ocean Proximity') # Title
for p in ax.patches: # Looping thru the records to get the values of bar
    ax.annotate(np.round(p.get_height(),decimals=0), (p.get_x()+p.get_width()/2.,   p.get_height()), ha='center', va='center', xytext=(2, 10), textcoords='offset points')
plt.tight_layout()
plt.show() # Displaying bar graph
```

```python
print ('Problem 2 Section 10\n')
print ('Scatter plot of age (x-axis) and population (y-axis)')
print ('Color based on price \n')
plt.scatter(readcsv['housing_median_age'], readcsv['population'], c=readcsv['median_house_value']) # Scatter with inputs as folows x value, y value and color based on column.
plt.xlabel("Age") # Naming X axis
plt.ylabel("Population")# Naming Y axis
plt.title('Age Vs Population') # Title
cbar= plt.colorbar() # Creating a color bar as per value
cbar.set_label("House Value ($)", labelpad=+1) # Naming the color bar
plt.show()# Displaying Scatter plot
```
