## Problem 4

## Problem 5
Min Max normalisation of "Income", "Age", "Children" attribute using the following formula.
Yi = [Xi - min(X)]/[max(X) - min(X)]

```python
income_min = vstable["income"].min()
income_max = vstable["income"].max()
income_norm = (vstable["income"] - income_min) / (income_max-income_min)
print("Min Max normalisation of column 'Income' :")
income_norm.head(10)
```

The above cell shows the Min Max normalisation of "Income".

```python
age_min = vstable["age"].min()
age_max = vstable["age"].max()
age_norm = (vstable["age"] - age_min) / (age_max-age_min)
print("Min Max normalisation of column 'Age' :")
age_norm.head(10)
```

The above cell shows the Min Max normalisation of "Age".

```python
children_min = vstable["children"].min()
children_max = vstable["children"].max()
children_norm = (vstable["children"] - children_min) / (children_max-children_min)
print("Min Max normalisation of column 'Children' :")
children_norm.head(10)
```

The above cell shows the Min Max normalisation of "Children".

```python
vstable2 = vstable.copy()
vstable2["age"] = age_norm
vstable2["children"] = children_norm
vstable2["income"] = income_norm

vstable2.head(10)
```

Replacing the normalised values of "Income", "Age", "Children" attribute in the dataframe and showing the first 10 records.
