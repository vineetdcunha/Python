The search function takes a query object (in this case a vector of words), and searches for the K most similar (least distant) items in the data (our index of documents). The "measure" parameter allows us to use either the Euclidean distance or the inverse of Cosine similarity as our ditance metric. The function returns the indices of the K most similar neighbors and a list of their distances to the query object.
#### Function to calculate distance

```python
def knn_search(x, D, K, measure):
    """ find K nearest neighbors of an instance x among the instances in D """
    if measure == 0:
        # euclidean distances from the other points
        dists = np.sqrt(((D - x)**2).sum(axis=1))
    elif measure == 1:
        # first find the vector norm for each instance in D as wel as the norm for vector x
        D_norm = np.array([np.linalg.norm(D[i]) for i in range(len(D))])
        x_norm = np.linalg.norm(x)
        # Compute Cosine: divide the dot product o x and each instance in D by the product of the two norms
        sims = np.dot(D,x)/(D_norm * x_norm)
        # The distance measure will be the inverse of Cosine similarity
        dists = 1 - sims
    idx = np.argsort(dists) # sorting
    # return the indexes of K nearest neighbors
    return idx[:K], dists
```

The function below will use our previous knn_search function to identify the label/category for for an instance x to be classifed, using the majority label of the K nearest neighbors of x.
#### Function to fetch first N neighbors

```python
def knn_classify(x, D, K, labels, measure):
    from collections import Counter
    neigh_idx, distances = knn_search(x, D, K, measure)
    neigh_labels = labels[neigh_idx]
    count = Counter(neigh_labels)
    #print("Labels for top ", K, "neighbors: ", count)
    return count.most_common(1)[0][0]
```

Using KNN classifer to evaluate it's classification accuracy. Running the classifier on each test instance in our test data and compare the predicted class to the actual class for each. We will maintain the number of agreements which allows us to compute the final accuracy rate across all test instances.
#### Evaluation

```python
def knn_evaluate(test_set, train_set, train_class, test_class, measure):
    numTestVecs = len(test_class)
    for k in range (5,101,5):
        accCount = 0.0
        for i in range(numTestVecs):
            classifierResult = knn_classify(test_set[i,:], train_set, k, train_class, measure)
            if (classifierResult == test_class[i]):
                accCount += 1.0
        lst.append([measure,k,accCount/float(numTestVecs)])
        print("Value of k:",k)
        print("Total accuracy rate:", accCount/float(numTestVecs) ,'\n')
```

Transforming the data to TFxIDF weights:

```python
# Note: doc frequency (df) for a term t is the number of docs in which t appears divided by total number of docs

# finding the doc counts for each term

DF = pd.DataFrame([(TD!=0).sum(1)]).T
DF
```

```python
# Create a matrix with all entries = NDocs
NMatrix=np.ones(np.shape(TD), dtype=float)*NDocs
np.set_printoptions(precision=2,suppress=True,linewidth=120)
print(NMatrix)
```

```python
# Convert each entry into IDF values
# IDF is the log of the inverse of document frequency
# IDF is only a function of the term, so all columns will be identical.

IDF = np.log2(np.divide(NMatrix, np.array(DF)))
```

```python
print(IDF)
```

```python
# Computing the TFxIDF values for each document-term entry
TD_tfidf = TD * IDF
```

Sum all the columns based on the classes. Create a training prototype.

```python
def train_prototype (train_set, grouping_column):
    vs_train_ptype = train_set.groupby(grouping_column).sum().reset_index()
    vs_target_train = vs_train_ptype[grouping_column]
    print('Output of training prototype with class variable')
    print(vs_train_ptype.head())
    del vs_train_ptype[grouping_column]
    DTM_R = np.array(vs_train_ptype)
    print('Separating the class variable')
    print(vs_target_train.head())
    vs_target_train = np.array(vs_target_train).ravel()
    return DTM_R, vs_target_train
```

Creating function to evaluate Rocchio method. We are using the classifer using the Cosine similarity and pick the most similar record.

The value of index is set to 1 because we want the most similar measure from prototype.

```python
def rocchio_measure(x, D):
    D_norm = np.array([np.linalg.norm(D[i]) for i in range(len(D))])
    x_norm = np.linalg.norm(x)
    # Compute Cosine: divide the dot product o x and each instance in D by the product of the two norms
    sims = np.dot(D,x)/(D_norm * x_norm)
    # The distance measure will be the inverse of Cosine similarity
    #dists = 1 - sims
    idx = np.argsort(-sims) # sorting
    return idx[:1]
       
```

```python
def rocchio_evaluate(test_set, train_set, train_class, test_class):
    numTestVecs = len(test_class)
    accCount = 0.0
    for i in range(numTestVecs):
        index = rocchio_measure(test_set[i,:], train_set) # Setting the value of k to 1 and reusing the previous classifier model
        classifierResult = train_class[index]
        if (classifierResult == test_class[i]):
            accCount += 1.0
    print("Total accuracy rate:", accCount/float(numTestVecs) ,'\n')
```
